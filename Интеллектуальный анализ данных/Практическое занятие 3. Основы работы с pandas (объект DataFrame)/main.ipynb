2. Изменить позиции столбцов объекта DataFrame (создать свой объект DataFrame).

- a) Поменять местами столбцы 'a' и 'c'

```python
new_df = pd.DataFrame(np.random.randint(0, 100, size=(4, 4)), columns = list('abcd'))
new_df
```

```python
new_df = new_df.reindex(columns=list('cbad'))
new_df
```

- b) Написать функцию, которая меняет столбцы местами

```python
def swap_col(df, col1, col2):
    new_columns = []
    for col in df.columns:
        if col == col1:
            new_columns.append(col2)
        else:
            if col == col2:
                new_columns.append(col1)
            else:
                new_columns.append(col)
    return df.reindex(columns=new_columns)

new_df = swap_col(new_df, 'b', 'd')
new_df
```

- c) Сортировать столбцы по наименованию

```python
new_df = new_df[np.sort(new_df.columns)]
new_df
```
