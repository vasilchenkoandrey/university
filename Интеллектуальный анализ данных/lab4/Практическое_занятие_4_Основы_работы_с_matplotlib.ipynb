# Практическое занятие 4. Основы работы с matplotlib
1. Построить график функции y=sin(x)

```python
import matplotlib.pyplot as plt
from math import sin
import numpy as np

x = np.arange(0, 7, 0.1)
y = [sin(i) for i in x]

plt.plot(x, y)
```

2.Нарисуйте круговую диаграмму и гистограмму, демонстрирующие диаметры планет (справочная информация: http://www.webteka.com/solarsystem-diameters-planets/) 

```python
planet = {'Меркурий': 4879,
          'Венера': 12103,
          'Земля': 12742,
          'Марс': 6780,
          'Юпитер': 139822,
          'Сатурн': 116464,
          'Уран': 50724,
          'Нептун': 49244}

names = [i for i in planet.keys()] 
diameter = [i for i in planet.values()]
```

```python
print('Гистограмма демонстрирующия диаметры планет')

plt.bar(names, diameter)
plt.show()
```

```python
print('Круговая диаграмма демонстрирующия диаметры планет')

plt.pie(diameter, labels = names)
plt.show()
```

3. Прочитать файл hubble.txt с данными измерений космического телескопа Хаббл и нарисовать диаграмму зависимости скорости убегания от объекта измерения.
  
  hubble.txt: https://moodle.cfuv.ru/pluginfile.php/401617/mod_assign/introattachment/0/hubble.txt?forcedownload=1

```python
import pandas as pd

df = pd.read_csv('hubble.txt')
x = df['distance']
y = df['recession_velocity']

print('Диаграмма зависимости скорости убегания от объекта измерения.')

plt.xlabel('distance')
plt.ylabel('recession_velocity')
plt.plot(x, y)
plt.show()
```
