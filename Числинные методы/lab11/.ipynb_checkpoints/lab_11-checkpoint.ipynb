Требуется найти приближенное решение уравнения: $y'=f(x,y), \ x∈[x_0,X],$
при заданном начальном условии: $y(x_0)=y_0.$ 

```python
from numpy import array, sin, e
def dydx(x, y):
    return (1 + sin(2*x) * e**(-x * (1 + x)) + 2 * (2 - x) * y) 

def rungeKutta(x0, y0, X, h):
    n = (int)((X - x0)/h)
    y = y0

    
    arrx0[0] = x0
    arry[0] = y
    
    print("{:<4} {:<6} {:<11} {:<11} {:<11} {:<11} {:<11}" .format("n","x0","y","k1","k2","k3","k4"))
    for i in range(0, n):
        k1 = h * dydx(x0, y)
        arrk1[i] = k1
        k2 = h * dydx(x0 + 0.5 * h, y + 0.5 * k1)
        arrk2[i] = k2
        k3 = h * dydx(x0 + 0.5 * h, y + 0.5 * k2)
        arrk3[i] = k3
        k4 = h * dydx(x0 + h, y + k3)
        arrk4[i] = k4
        print("{:<4d} {:<6.1f} {:<11f} {:<11f} {:<11f} {:<11f} {:<11f}" .format(i,x0,y,k1,k2,k3,k4))
        y = y + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
        x0 = x0 + h
        arrx0[i] = x0
        arry[i] = y
    #return y


x0 = 1
y = 10
X = 6
h = 0.1
n = (int)((X - x0)/h)
arrx0 = zeros(n)
arry = zeros(n)
arrk1 = zeros(n)
arrk2 = zeros(n)
arrk3 = zeros(n)
arrk4 = zeros(n)
```

$$f(x,y) = (a + b(x)) \cdot e^{c(x)} + 2 \cdot (2 - x) \cdot y$$
$$a=1$$
$$b=sin(2x)$$
$$c=-x \cdot (1 + x)$$
$$h=0.1$$


```python
rungeKutta(x0, y, X, h)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(arrx0, arrk1, arrx0, arrk2, arrx0, arrk3, arrx0, arrk4)
plt.legend(['k1', 'k2', 'k3', 'k4'], loc='best')
plt.grid(True)
```

```python
x = range(0, n, 1)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, arry, arrk1, x, arry, arrk2, x, arry, arrk3, x, arry, arrk4, label='parametric curve', linewidth=2.0)
```
